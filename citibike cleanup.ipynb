```python
import pandas as pd
import glob
import os
```

```python
# set search path and find all csv files
path = 'Resources'
all_files = glob.glob(path + "/*.csv")
```

```python
# create empty list to store dataframes
li = []

for f in all_files:
    #read in csv
    df_raw = pd.read_csv(f, index_col=None)
    df_raw['filename'] = os.path.basename(f)
    li.append(df_raw)

df = pd.concat(li,axis=0,ignore_index=True)
print(df.shape)
df.head()
```

```python
df.count()
```

```python
# drop rows with NAN data
df = df.dropna()
df.count()
```

```python
# convert tripduration from seconds to minutes
df['tripduration'].apply(lambda x: pd.to_datetime(x, unit='s').dt.strftime('%H:%M:%S'))
df.head()
```

```python
# split starttime and stoptime
df['startdate'] = pd.to_datetime(df['starttime']).dt.date
df['start_time'] = pd.to_datetime(df['starttime']).dt.time

df['stopdate'] = pd.to_datetime(df['stoptime']).dt.date
df['stop_time'] = pd.to_datetime(df['stoptime']).dt.time

df.head()
```

```python
# replace numbers with genders
df.gender = df.gender.map({1:"Male", 2:"Female", 0:"Unknown"})
df.head()
```

```python
# create an age column
#create year column from starttime
df['year']=pd.DatetimeIndex(df['starttime']).year

df['age']= 2018 - df['birth_year']
```

```python
#drop starttime and stoptime columns
df.drop(columns=["stoptime"], inplace=True)
df.head()
```

```python
# add year start date column
df.loc[:'year']=df['startdate']
```

```python
# save dataset as csv
df.to_csv("Resources/2017_2018_tripdata_JC")
```
