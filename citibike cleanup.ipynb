```python
import pandas as pd
import glob
import os
```

```python
# set search path and find all csv files
path = 'Resources'
all_files = glob.glob(path + "/*.csv")
```

```python
# create empty list to store dataframes
li = []

for f in all_files:
    #read in csv
    df_raw = pd.read_csv(f, index_col=None)
    df_raw['filename'] = os.path.basename(f)
    li.append(df_raw)

df = pd.concat(li,axis=0,ignore_index=True)
print(df.shape)
df.head()
```

```python
df.count()
```

```python
# drop rows with NAN data
df = df.dropna()
df.count()
```

```python
# filter out trips under 3600 seconds
df.loc[df['tripduration']>3600]
```

```python
#df['tripduration'] = pd.to_datetime(df['tripduration'], unit='s')
```

```python
# calculate age of riders
ages = []

for i in df["birth year"]:
    age = 2019 - i
    ages.append(age)

df["age"] = ages
df.head()
```

```python
# save dataset as csv
df.to_csv("Resources/2019_tripdata_JC.csv", index=False)
```
